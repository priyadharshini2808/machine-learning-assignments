```python
book_series  = pd.DataFrame(pd.Series(book_list))
```

```python
book_series
```

```python
book_series = book_series.iloc[:2001,:]
```

```python
book_series.columns = ["transactions"]
```

```python
X = book_series['transactions'].str.join(sep='*').str.get_dummies(sep='*')
```

```python
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
```

```python
plt.xlabel('frequent_itemsets');plt.ylabel('support')
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules
```

```python
rules.sort_values('lift',ascending = False,inplace=True)
```

```python
def to_list(i):
    return (sorted(list(i)))
```

```python
ma_X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)
```

```python
ma_X = ma_X.apply(sorted)
```

```python
rules_sets = list(ma_X)
```

```python
unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules_sets)]
```

```python
index_rules = []
```

```python
for i in unique_rules_sets:
    index_rules.append(rules_sets.index(i))
```

```python
rules_no_redudancy  = rules.iloc[index_rules,:]
```

```python
rules_no_redudancy.sort_values('lift',ascending=False)
```
